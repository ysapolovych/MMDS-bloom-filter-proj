```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_pickle('data/train.pkl')
```

```python
changes_by_type = data.groupby(['bot', 'type']).size().reset_index().rename(columns={0: 'count'})
```

```python
px.bar(changes_by_type, x='type', y='count', color='bot',
       title='Distribution of changes by type and user category',
       width=600, height=500,
       labels={'bot': 'User is bot', 'type': 'Change type', 'count': 'Count of changes'},
       color_discrete_sequence=px.colors.qualitative.Pastel)
```

```python
data['change_length'] = data['length.new'] - data['length.old']
```

```python
fig = px.histogram(data[~data['change_length'].isna()], x='change_length', color='bot', marginal="rug",
                   title='Distribution of edits lengths by user category',
                   width=1200, height=600,
                   labels={'bot':'User is bot', 'change_length': 'Change length', 'count': 'Count of changes'},
                   color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(opacity=0.75)
fig.show()
```
